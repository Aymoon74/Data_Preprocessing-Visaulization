## Detect + Handle Outliers

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

for col in df.select_dtypes(include=np.number).columns:
  df = remove_outliers_iqr(df, col)
```

## Visualization

```python
oldest_players = df.sort_values(by='age', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='short_name', y='age', data=oldest_players, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Player Name')
plt.ylabel('Age')
plt.title('Ages of the 10 Oldest Players')
plt.tight_layout()
plt.show()
```

```python
youngest_players = df.sort_values(by='age', ascending=True).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='short_name', y='age', data=youngest_players, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Player Name')
plt.ylabel('Age')
plt.title('Ages of the 10 youngest Players')
plt.tight_layout()
plt.show()
```

```python
largest_Value = df.nlargest(10, 'value_eur')
plt.figure(figsize=(8, 6))
plt.scatter(largest_Value['short_name'], largest_Value['value_eur'])
plt.xticks(rotation=90)
plt.xlabel('Player Name')
plt.ylabel('value_eur (EUR)')
plt.title('10 largest value_eur')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('players_20.csv')

netherlands_players = df[df['nationality'] == 'Netherlands']

players_data = netherlands_players[['short_name', 'player_positions', 'overall']]

greatest_players = players_data.sort_values('overall', ascending=False).head(10)


plt.figure(figsize=(10, 6))
for index, row in greatest_players.iterrows():
    plt.text(row['overall'], row['short_name'], row['player_positions'], ha='left', va='center')

plt.scatter(greatest_players['overall'], greatest_players['short_name'])
plt.xlabel('Overall Rating')
plt.ylabel('Player Name')
plt.title('Top 10 Netherlands Players by Overall Rating (with Positions)')
plt.grid(True)
plt.show()
```

```python
germany_players = df[df['nationality'] == 'Germany']

players_data = germany_players[['short_name', 'player_positions', 'overall']]

greatest_players = players_data.sort_values('overall', ascending=False).head(10)

plt.figure(figsize=(10, 6))
for index, row in greatest_players.iterrows():
    plt.text(row['overall'], row['short_name'], row['player_positions'], ha='left', va='center')

plt.scatter(greatest_players['overall'], greatest_players['short_name'])
plt.xlabel('Overall Rating')
plt.ylabel('Player Name')
plt.title('Top 10 Germany Players by Overall Rating (with Positions)')
plt.grid(True)
plt.show()
```

```python
foot_counts = df['preferred_foot'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(foot_counts, labels=foot_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Preferred Foot Among Players')
plt.axis('equal')
plt.show()
```

```python
club_value = df.groupby('club')['value_eur'].sum()

top_10_clubs = club_value.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_10_clubs.plot(kind='bar', color='skyblue')
plt.title('Top 10 Clubs by Total Player Value')
plt.xlabel('Club Name')
plt.ylabel('Total Value (€)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
